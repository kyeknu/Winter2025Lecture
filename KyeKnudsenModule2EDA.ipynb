```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
#df=pd.read_csv(r'C:\Users\kyekn\Desktop\Data Analytics\Advanced Python')
df=pd.read_csv('housing.csv')
```

```python
df
```

```python
sns.scatterplot(data=df, x='Landsize', y='Price', hue='Type')
```

```python
Avg_Price_Per_Year=df.groupby('YearBuilt').agg({'Price': 'mean'})
```

```python
Avg_Price_Per_Year
```

```python
sns.lineplot(x='YearBuilt', y='Price', data=Avg_Price_Per_Year)
```

```python
Region_Price=df.groupby('Regionname').agg({'Price': 'mean'})
```

```python
Region_Price
```

```python
sns.barplot(x='Regionname', y='Price', data=Region_Price)
```

```python
sns.histplot(x='Price', data=df)
```

```python
sns.boxplot(x='YearBuilt', y='Regionname', data=df)
```
