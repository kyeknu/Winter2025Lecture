```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df=pd.read_csv('companies.csv')
```

```python
df.head()
```

```python
sns.countplot(x='state', data=df)
plt.show()
```

```python
df_job_openings=df.groupby('Industry', as_index=False).agg({'job_openings': 'sum'})
```

```python
sns.barplot(x='job_openings', y='Industry', data=df_job_openings)
```

```python
df_founded_mean=df.groupby('founded', as_index=False).agg({'growth_percentage': 'mean'})
```

```python
sns.lineplot(x='founded', y='growth_percentage', data=df_founded_mean)
```

```python
sns.scatterplot(x='employees', y='estimated_revenues', data=df)
plt.show()
```

```python
sns.histplot(x='founded', data=df)
```
